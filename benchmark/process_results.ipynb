```python
import json
import ast
```

```python
def process_execution_stats(jdata):
    for task in jdata:
        statstops = jdata[task]['msg']['startstops']
        output_size = jdata[task]['msg']['nbytes']
        input_size = 0

        comp_start_time = 0
        comp_end_time = 0

        comm_start_time = 0
        comm_end_time = 0
        for ss in statstops:
            if ss['action'] == 'compute':
                if not comp_start_time or (comp_start_time and comp_start_time >  ss['start']):
                    comp_start_time = ss['start']
                if not comp_end_time or (comp_end_time and comp_end_time <  ss['stop']):
                    comp_end_time = ss['stop']


            elif ss['action'] == 'transfer':
                if not comm_start_time or (comm_start_time and comm_start_time >  ss['start']):
                    comm_start_time = ss['start']
                if not comm_end_time or (comm_end_time and comm_end_time <  ss['stop']):
                    comm_end_time = ss['stop']
                input_size += ss['nbytes']


        print(f'{task}\n\t compute time: {round(comp_end_time - comp_start_time, 2)} \
        \n\t communication time: {round(comm_end_time - comm_start_time, 2)} \
        \n\t input size: {input_size} \n\t output size: {output_size} \n\n')

        #break
        
```

```python
with open('/opt/dask-distributed/benchmark/svd_tall_skinny_matrix_200k_x_1k.json', 'r') as fd:
    raw = fd.read()
    jdata = ast.literal_eval(raw)
    process_execution_stats(jdata)
```

```python
with open('/opt/dask-distributed/benchmark/svd_tall_skinny_matrix_1M_x_1k.json', 'r') as fd:
    raw = fd.read()
    jdata = ast.literal_eval(raw)
    process_execution_stats(jdata)
```
